```python
#Import Library

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
def plot_shape(data):
    """
    Plots the shape of a dataset in 2D with color coding based on class labels.

    Parameters:
    data (numpy.ndarray): A 2D array where the first two columns are features 
                          and the third column is the class label.
    """

    # Extract features (first two columns) and labels (third column)
    X = data[:, :2]  # Feature matrix
    y = data[:, 2]   # Class labels

    # Define a custom colormap for the labels
    cmap_bold = ListedColormap(['#0000FF', '#FFD700'])

    # Scatter plot of points with colors based on labels
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold)

    # Determine axis limits with padding for better visualization
    plt.xlim(X[:, 0].min() - 1, X[:, 0].max() + 1)  # Set x-axis limits
    plt.ylim(X[:, 1].min() - 1, X[:, 1].max() + 1)  # Set y-axis limits

    # Set the title of the plot
    plt.title('Shape of the dataset')

    # Display the plot
    plt.show()
```

```python
data1 = np.genfromtxt('/kaggle/input/coordinate-based-shapes-and-level-dataset/1.ushape.csv', delimiter=',')
plot_shape(data1)
```
